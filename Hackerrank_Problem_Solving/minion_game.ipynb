```python
# checking character is vowel or consonant
def is_vowel(ch):
    if(ch.lower() in ['a','e','i','o','u']):
        return True
    else:
        return False
    
```

```python
# counting words start with vowels
def team_vowel(word):
    vowel_words = []
    for i in range(len(word)):
        if(is_vowel(word[i])):
            for j in range(len(word)-i):
                vowel_words.append(word[i:(len(word)-j)])      
        else:
            i+=1
    return len(vowel_words)

 
# counting words start with consonants 
def team_consonant(word):
    cons_words = []
    for i in range(len(word)):
        if(is_vowel(word[i])==False):
            for j in range(len(word)-i):
                cons_words.append(word[i:(len(word)-j)])      
        else:
            i+=1
    return len(cons_words)
        
    
print(team_vowel(s))
print(team_consonant(s))
```
